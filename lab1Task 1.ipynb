```python
import numpy as np
import matplotlib

# Creating a random 4x4 array from an exponential distribution
exp_array_4x4 = np.random.exponential(scale=1.0, size=(4, 4))
print("4x4 Exponential Distribution Array:")
print(exp_array_4x4)
```

```python
import matplotlib.pyplot as plt

# Creating a random 100000x1 array from an exponential distribution
exp_array_100000 = np.random.exponential(scale=1.0, size=100000)

# Creating uniform and normal distribution arrays for comparison
uniform_array = np.random.rand(100000)
normal_array = np.random.randn(100000)

# Plotting the histograms
plt.figure(figsize=(10, 6))
plt.hist(exp_array_100000, density=True, bins=100, histtype="step", color="red", label="Exponential")
plt.hist(uniform_array, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_array, density=True, bins=100, histtype="step", color="green", label="Normal")
plt.axis([-2.5, 10, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Creating X and Y data
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# Plotting the 3D surface
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap=matplotlib.cm.coolwarm, linewidth=0.1)

# Adding labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('$Z = X^2 + Y^2$')

plt.show()
```
